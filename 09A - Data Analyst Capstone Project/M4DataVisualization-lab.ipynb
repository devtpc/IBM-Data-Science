# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

q= "SELECT ConvertedComp FROM master"
df_cc = pd.read_sql_query(q, conn)
sns.histplot(df_cc["ConvertedComp"])
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here

q= "SELECT Age FROM master"
df_age = pd.read_sql_query(q, conn)
sns.boxplot(df_age["Age"])
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
q= "SELECT Age, WorkWeekHrs FROM master"
df_scatter = pd.read_sql_query(q, conn)
sns.scatterplot(df_scatter, x='Age', y='WorkWeekHrs')
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
q= "SELECT Age, WorkWeekHrs, CodeRevHrs FROM master"
df_bubble = pd.read_sql_query(q, conn)
sns.scatterplot(df_bubble, x='CodeRevHrs', y='WorkWeekHrs', size='Age')
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here

table_name = 'DatabaseDesireNextYear'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

```python
q= """SELECT DatabaseDesireNextYear, COUNT(DatabaseDesireNextYear) as count 
FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY count DESC
LIMIT 5
"""
df_pie = pd.read_sql_query(q, conn)
plt.pie(x=df_pie['count'], labels = df_pie['DatabaseDesireNextYear'], autopct='%1.1f%%')
```

```python
# Additional assignments
# Question: In the list of most popular languages respondents wish to learn next year, what is the rank of Python?

table_name = 'LanguageDesireNextYear'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])

```

```python
# question: How many respondents indicated that they currently work with 'SQL'? 
table_name = 'LanguageWorkedWith'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

```python
# question: How many respondents indicated that they currently work with 'SQL'? 
q= """SELECT LanguageWorkedWith, COUNT(LanguageWorkedWith) as count 
FROM LanguageWorkedWith
WHERE LanguageWorkedWith='SQL'
GROUP BY LanguageWorkedWith
ORDER BY count DESC
"""
df_sql= pd.read_sql_query(q, conn)
df_sql
```

```python
# question: How many respondents indicated that they work on 'MySQL' only? 
q= """SELECT COUNT(Respondent)
FROM 
(
SELECT Respondent, COUNT(DatabaseWorkedWith) count2, MAX(DatabaseWorkedWith) db FROM DatabaseWorkedWith
GROUP BY Respondent
HAVING count2=1
)
WHERE db='MySQL'

"""
df_db= pd.read_sql_query(q, conn)
df_db
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
q= "SELECT Age, WorkWeekHrs, CodeRevHrs FROM master WHERE Age BETWEEN 30 AND 35"
df_stacked = pd.read_sql_query(q, conn)
df_medians = df_stacked.groupby('Age')['WorkWeekHrs', 'CodeRevHrs'].median()
df_medians.plot(kind='bar', stacked=True)
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
q= "SELECT Age, ConvertedComp FROM master WHERE Age BETWEEN 45 AND 60"
df_line = pd.read_sql_query(q, conn)
df_line_medians = df_line.groupby('Age')['ConvertedComp'].median()
print(df_line_medians)
df_line_medians.plot(kind='line')
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
q= "SELECT MainBranch, COUNT(MainBranch) FROM master GROUP BY MainBranch"
df_mb = pd.read_sql_query(q, conn).set_index('MainBranch')
print(df_mb)
df_mb.plot(kind='barh')
```
